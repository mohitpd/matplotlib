```python
import matplotlib.pyplot as plt
x1 = [1,2,3,4,5,6]
y1 = [6,8,7,6,4,9]
x2 = [1,2,3,4,5,6,7]
y2 = [4,5,9,2,1,1,2]
plt.plot(x1, y1, label = 'Output 1')
plt.plot(x2, y2, label = 'Output 2')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Random Graph')
plt.legend()

plt.show()
```

```python
x1 = [2,4,6,8,10]
y1 = [5,2,1,7,9]
x2 = [1,3,5,7,9]
y2 = [4,7,9,3,1]
plt.bar(x1, y1, label = 'First Bars', color = 'r')
plt.bar(x2, y2, label = 'Second Bars', color = 'b')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.show()
```
